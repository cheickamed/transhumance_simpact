## libraries importation

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
sns.set_style("darkgrid")
```

## Spatial distribution in full space for unconstrained herd movements


```python
### 10 élements de réseaux sociales ; jour soc=0
#--------------------------  chargement des données ------------------------------
data=pd.read_csv('data/temoin_aller_50r.csv')

v1=data['grid_x']
v2=data['grid_y']
#v=data['nb_trp_inside']
#N=data['impact_trp_aller']
h=data['s_nb_trp_inside']
#print(len(h))
datar=pd.read_csv('data/temoin_retour_50r.csv')

v1r=datar['grid_x']
v2r=datar['grid_y']
#vr=datar['nb_trp_inside']
#Nr=datar['impact_trp_retour']
hr=datar['s_nb_trp_inside']

##-------------------------------------------------Distribution spatiale du trp en fonction la vegetation -----------
#--------aller---------
k=0
m=np.full((38,36),0) 
while k<len(v1):
    for i in range(0,38):
        for j in range(0,36):
            m[i][j]=m[i][j]+h[k]
            k=k+1

#print(m)
#print(k)
#print(m[m!=0.0])
v14=np.array([i for i in range(1,39)]) 
v15=np.array([i for i in range(1,37)])
data2=pd.DataFrame(m,v14,v15)
#---- retour--------
kr=0
mr=np.full((38,36),0) 
#print(mr)
while kr<len(v1r):
    for i in range(0,38):
        for j in range(0,36):
            mr[i][j]=mr[i][j]+hr[kr]
            kr=kr+1

data2r=pd.DataFrame(mr,v14,v15)


plt.figure(figsize=(8,4),dpi=200)
plt.subplot(1,2,1)
plt.title('Outward')# \n Duration: $ 26\pm 1$ days \n impacted cells: {:.2f}% $ \pm 0.9$" .format(len(h[h!=0])/len(h)*100))#$\pm $ jours \n
sns.heatmap(data2,cmap='Oranges_r',xticklabels=False,yticklabels=False,vmin=800.0,vmax=2000,robust=True,cbar=False)#
plt.subplot(1,2,2)
plt.title('Return')# \n Duration: $ 22 \pm 1$ days \n impacted cells {:.2f}% $\pm 0.94$'.format(len(hr[hr!=0])/len(hr)*100))
sns.heatmap(data2r,cmap='Oranges_r',xticklabels=False,yticklabels=False,vmin=800.0,vmax=2000,robust=True,cbar=False)

plt.savefig('Spatial_distribution_in_full_space_for_unconstrained_herd_movements')
plt.show()
```
